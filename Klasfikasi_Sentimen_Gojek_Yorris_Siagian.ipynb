```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder
```

#Load Data

```python
from google.colab import files
# Mengunggah file
uploaded = files.upload()
# Membaca file dengan encoding yang sesuai
for filename in uploaded.keys():
    df = pd.read_csv(filename, encoding="utf-8", keep_default_na=False)

# Menampilkan 5 baris pertama
df.head()
```

#Ekstraksi Fitur dengan TF-IDF

```python
# Ekstraksi Fitur dengan TF-IDF
X = df['text_akhir']
y = df['sentiment']

tfidf = TfidfVectorizer(min_df = 5,
                             max_df = 0.8,
                             sublinear_tf = True,
                             use_idf = True)
X_tfidf = tfidf.fit_transform(X)
```

```python
# Konversi hasil ekstraksi fitur menjadi dataframe
features_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.get_feature_names_out())

# Menampilkan hasil ekstraksi fitur
features_df
```

#3. Splitting Data

```python
# Membagi data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

# Label Encoding untuk Target (Sentimen)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
```

#Modeling

**MLP**

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# Membuat model MLP
mlp_model = MLPClassifier(
    hidden_layer_sizes=(512, 256, 128),
    activation='relu',
    solver='adam',
    alpha=1.0,
    learning_rate='adaptive',
    max_iter=1000,
    early_stopping=True,
    n_iter_no_change=10,
    random_state=42
)


# Melatih model
mlp_model.fit(X_train, y_train_enc)

# Prediksi untuk data training dan testing
y_pred_train_mlp = mlp_model.predict(X_train)
y_pred_test_mlp = mlp_model.predict(X_test)

# Evaluasi akurasi
accuracy_train_mlp = accuracy_score(y_train_enc, y_pred_train_mlp)
accuracy_test_mlp = accuracy_score(y_test_enc, y_pred_test_mlp)

# Menampilkan hasil evaluasi
print(f'Akurasi Training MLP: {accuracy_train_mlp:.4f}')
print(f'Akurasi Testing  MLP: {accuracy_test_mlp:.4f}')
print('\nClassification Report (Testing - MLP):\n', classification_report(y_test_enc, y_pred_test_mlp))
```
